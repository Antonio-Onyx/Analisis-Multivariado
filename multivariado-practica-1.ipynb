# Matriz de Datos

>  primero tenemos que generar una matriz de datos con una distribucion
$
N_{8} (\mu,\Sigma) 
$

```r
# instala el paquete MASS si no esta instalado
if(!require(MASS)){
    install.packages("MASS")
}
install.packages("tidyverse")
install.packages("viridis")
```

```r
library(ggplot2)
library(viridis)
```

```r
#cargamos la paqueteria
library(MASS)

#declaramos los parametros

mu <- rep(0,8) #es el vector de medias
sigma <- diag(8) #es la diagonal de la matriz de covarianzas, que es simetrica

set.seed(123) #seed nos genera numeros aleatorios o pseudoaleatorios
            #al utilizar seed(123) lo que hace es que establece seed en "123"
            #para que los numeros aleaotorios que obtengamos sean los mismos, aun si
            #los ejecutamos en un entorno diferento o equipo.

datos <- mvrnorm(100, mu=mu,Sigma=sigma)
```

```r
head(datos)
```

```r
df <- as.data.frame(datos)
#print(dt)
```

# Estadisticas descriptivas Multivariadas
* $\overline{x}$ es el vector de medias
* $S$ es la matriz de covarianzas
* $\tilde{x}$ es la mediana(?

```r
print("covariance matrix")
cov_matr <- cov(df)
print(cov_matr)
```

```r
heatmap(cov_matr, 
        col =viridis(100),
        main="Heatmap de la matriz de covarianza",
        xlab="X1",
        ylab="Y1",
        labRow=names(df),
        labCol=names(df),
        symm=TRUE)
```
